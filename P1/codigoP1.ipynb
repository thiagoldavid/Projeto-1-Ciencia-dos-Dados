***HIPÓTESE?***

Como se comportaria a inflação de um país com sua entrada na OPEP (Organização dos Países Exportadores de Petróleo), devido ao controle de produção e preços que são regulamentados para todos países da organização.

 ***MECANISMO?***
 
 Ao entrar na OPEP, o país passaria a seguir as regras adotadas por todos países membros, variando sua produção. O aumento dessa faria com que o preço do petróleo abaixasse ,e vice-versa, podendo aumentar ou diminuir a inflação.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
from numpy import arange
from numpy import percentile
import numpy as np
```

Criando a tabela com dados sobre a produção de petróleo dos países.

```python
oil_production = pd.read_excel('Oil Production.xls.xlsx')
oil_production.head()
```

Colocando como índice os nomes dos países.

```python
nomedospaises = oil_production.set_index('Crude oil production, total (toe)')
nomedospaises.head()

```

Especificando os países a serem analisados.

```python
dadosdaproducaoGERAL =nomedospaises.loc[['Algeria','China',"Colombia",'Ecuador','Egypt','Gabon','India','Iran','Kuwait','Mexico','Peru','Saudi Arabia','Trinidad and Tobago','Venezuela']]
dadosdaproducaoGERAL
```

Foram retirados alguns países membros atuais, visto que entraram na organização após o período que está sendo analisado ou não possuem dados concretos.

```python
dadosdaproducaoOPEP =nomedospaises.loc[['Algeria','Ecuador','Iran','Kuwait','Saudi Arabia','Trinidad and Tobago','Venezuela']]
dadosdaproducaoNAOOPEP =nomedospaises.loc[['China','Colombia','Egypt','Gabon','India','Mexico','Peru']]
```

Especificando os anos a serem analisados pelo projeto.

```python
dadosdaproducao =dadosdaproducaoGERAL.loc[:,[1965,1966,1967,1968,1969,1970,1971,1972,'OPEP']]
dadosdaproducao

```

Criando variáveis para plotar os gráficos.

```python
dadosdaproducao1 =dadosdaproducaoOPEP.loc[:,[1965,1966,1967,1968,1969,1970,1971,1972]]
dadosdaproducao2 =dadosdaproducaoNAOOPEP.loc[:,[1965,1966,1967,1968,1969,1970,1971,1972]]













```

Plotando os gráficos.

```python
dadosdaproducao1.plot.bar()
plt.title('Producao de oleo em paises membros da OPEP')
plt.xlabel('Países')
plt.ylabel('Quantidade produzida de oleo')
plt.show(True)

dadosdaproducao1.plot.box()
plt.title('Producao de oleo em paises membros da OPEP')
plt.xlabel('Anos')
plt.ylabel('Quantidade produzida de oleo')
plt.show(True)

dadosdaproducao2.plot.box()
plt.title('    Producao de oleo em paises nao membros da OPEP')
plt.xlabel('Anos')
plt.ylabel('Quantidade produzida de oleo')
plt.show(True)

dadosdaproducao2.plot.bar()
plt.title('    Producao de oleo em paises nao membros da OPEP')
plt.xlabel('Países')
plt.ylabel('Quantidade produzida de oleo')
plt.show(True)


```









Criando a tabela com dados sobre a inflação dos países.

```python
inflacao = pd.read_excel('Inflation.xlsx')
inflacao.head()
```

Colocando como índice os nomes dos países.

```python
nomedospaises2 = inflacao.set_index('Inflation, GDP deflator (annual %)')
nomedospaises2.head()

```

Especificando os países a serem analisados.

```python
dadosdainflacaogeral = nomedospaises2.loc[['Algeria','China',"Colombia",'Ecuador','Egypt','Gabon','India','Iran','Kuwait','Mexico','Peru','Saudi Arabia','Trinidad and Tobago','Venezuela']]
dadosdainflacaogeral
```

Foram retirados alguns países membros atuais, visto que entraram na organização após o período que está sendo analisado ou não possuem dados concretos.

```python


dadosdainflacaoOPEP =dadosdainflacaogeral.loc[['Algeria','Ecuador','Iran','Kuwait','Saudi Arabia','Trinidad and Tobago','Venezuela']]
dadosdainflacaoNAOOPEP =dadosdainflacaogeral.loc[['China','Colombia','Egypt','Gabon','India','Mexico','Peru']]



```

Especificando os anos a serem analisados pelo projeto.

```python
dadosdainflacao =dadosdainflacaogeral.loc[:,['1965','1966','1967','1968','1969','1970','1971','1972','OPEP']]
dadosdainflacao


```

Criando variáveis para plotar os gráficos.

```python
dadosdainflacao1 =dadosdainflacaoOPEP.loc[:,['1965','1966','1967','1968','1969','1970','1971','1972']]
dadosdainflacao2 =dadosdainflacaoNAOOPEP.loc[:,['1965','1966','1967','1968','1969','1970','1971','1972']]
```

Plotando os gráficos.

```python
dadosdainflacao1.plot.box()
plt.title('Inflação nos países membros da OPEP')
plt.xlabel('Anos')
plt.ylabel('Inflação(%)')
plt.show(True)


dadosdainflacao1.plot.bar()
plt.title('Inflação nos países membros da OPEP')
plt.xlabel('Países')
plt.ylabel('Inflação(%)')
plt.grid()
plt.show(True)



dadosdainflacao2.plot.box()
plt.title('Inflação nos países não membros da OPEP')
plt.xlabel('Anos')
plt.ylabel('Inflação(%)')
plt.show(True)


dadosdainflacao2.plot.bar()
plt.title('Inflação nos países não membros da OPEP')
plt.xlabel('Países')
plt.ylabel('Inflação(%)')
plt.grid()
plt.show(True)



```

Agora vamos plotar um gráfico de dispersão para ver a correlação entre a produção de petróleo e da inflação dos países membros da OPEP.

```python
inflacao_1968=dadosdainflacaoOPEP.loc[:, ['1968']]
prodoil_1968=dadosdaproducaoOPEP.loc[:, [1968]]



dados= prodoil_1968.join(inflacao_1968, how='inner', lsuffix='oil_', rsuffix='_inflacao')
plot=dados.plot.scatter(x=1968,y='1968')
plt.grid()
plt.title('Produção de Petróleo x Inflação em 1968 dos países membros da OPEP')
plt.ylabel('Inflação(%)')
plt.xlabel('Produção de Petróleo')
print('A correlação entre a inflação em 1968 e a produção de petróleo dos países membros da OPEP é:', dados[1968].corr(dados['1968']))



```

Agora vamos plotar um gráfico de dispersão para ver a correlação entre a produção de petróleo e da inflação dos países que estão fora da OPEP.

```python
inflacao_1968=dadosdainflacaoNAOOPEP.loc[:, ['1968']]
prodoil_1968=dadosdaproducaoNAOOPEP.loc[:, [1968]]


dados= prodoil_1968.join(inflacao_1968, how='inner', lsuffix='oil_', rsuffix='_inflacao')
plot=dados.plot.scatter(x=1968,y='1968')
plt.grid()
plt.title('Produção de Petróleo x Inflação em 1968 dos países que estão fora da OPEP')
plt.ylabel('Inflação(%)')
plt.xlabel('Produção de Petróleo')
print('A correlação entre a inflação em 1968 e a produção de petróleo dos países que estão fora da OPEP é:', dados[1968].corr(dados['1968']))

```

    Por fim, plotaremos um gráfico de dispersão para ver a correlação entre a produção de petróleo e da inflação  de todos os países analisados

```python
inflacao_1968=dadosdainflacaogeral.loc[:, ['1968']]
prodoil_1968=dadosdaproducaoGERAL.loc[:, [1968]]


dados= prodoil_1968.join(inflacao_1968, how='inner', lsuffix='oil_', rsuffix='_inflacao')
plot=dados.plot.scatter(x=1968,y='1968')
plt.grid()
plt.title('Produção de Petróleo x Inflação em 1968 dos países que estão fora da OPEP')
plt.ylabel('Inflação(%)')
plt.xlabel('Produção de Petróleo')
print('A correlação entre a inflação em 1968 e a produção de petróleo é:', dados[1968].corr(dados['1968']))
```

***CONCLUSÃO***

   Pode-se concluir, portanto, a partir dos dados e gráficos analisados, que a produção de petróleo não influencia diretamente na inflação do mesmo país e que existem outros diversos fatores que possam interferir nessa relação. Isso foi evidente ao nos depararmos com a grande dispersão dos pontos tanto dos gráficos de dispersão dos países membros da OPEP tanto quanto os do que estão fora da organização, demonstrando a relação direta inexistente entre a produção de petróleo e a inflação dos países analisados.
